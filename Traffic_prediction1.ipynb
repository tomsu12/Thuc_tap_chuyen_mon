# 1. Introduction: giới thiệu

Dự đoán lưu lượng giao thông (TFP) có nghĩa là dự đoán khối lượng và mật độ lưu lượng giao thông, thường là để kiểm soát chuyển động của xe, giảm ùn tắc giao thông và tạo tuyến đường tối ưu (ít thời gian nhất hoặc tiêu tốn năng lượng). Với sự tiến bộ gần đây trong Trí tuệ nhân tạo, Học máy (ML), Học sâu (DL) và Dữ liệu lớn, nghiên cứu trong lĩnh vực dự đoán lưu lượng giao thông đã được mở rộng rộng rãi.  
TFP là thành phần chính của Hệ thống giao thông thông minh (ITS) và có thể hỗ trợ ITS dự báo lưu lượng giao thông. Các thành phố lớn có quy định giao thông cực kỳ khó khăn. nhiều quốc gia đã áp dụng ITS để giảm chi phí liên quan đến tắc nghẽn giao thông. Nghiên cứu này xem xét việc áp dụng mạng nơ-ron nhân tạo (ANN), ML, DL và các kỹ thuật và mô hình khác cho TFP. Cuối cùng, chúng tôi sẽ đề xuất mô hình dự đoán của riêng mình bằng DL, đào tạo và kiểm tra nó, phân tích độ chính xác và so sánh độ chính xác của mô hình của chúng tôi với các mô hình khác.  
So sánh với các phương pháp ML thông thường, các mô hình DL có những ưu điểm như đơn giản hóa các quy trình tiền xử lý dữ liệu và vượt trội hơn các phương pháp ML khác về độ chính xác. Do đó, dự đoán lưu lượng truy cập theo hướng dữ liệu do sự sẵn có của dữ liệu lưu lượng lớn và sơ đồ DL do các thủ tục tiền xử lý dữ liệu đã nhận được sự chú ý rộng rãi gần đây trong TFP.  
Hơn nữa, trong những năm gần đây đã có sự gia tăng lớn về dữ liệu có sẵn với sự tiến bộ của các thành phố thông minh. Việc hiện đại hóa này có thể có tác động thuận lợi đến mạng lưới giao thông trong khu vực ITS, giảm thời gian đi lại, tăng năng suất và giảm thiểu tác động môi trường của phương tiện. Công nghệ ML và DL là những lĩnh vực phát triển nhanh để dự đoán lưu lượng truy cập. Tín hiệu giao thông, tai nạn, điều kiện thời tiết và sửa chữa đường là nguyên nhân chính của giao thông. Vì dữ liệu lưu lượng truy cập thời gian thực phần lớn được tạo ra theo cấp số nhân, các nguyên tắc dữ liệu lớn phải được sử dụng để cải thiện việc vận chuyển dữ liệu. Thực tế này thúc đẩy chúng tôi dự đoán lưu lượng giao thông giữa Minneapolis và St. Paul tại một điểm cụ thể ở Minnesota. Mục tiêu của chúng tôi là xây dựng một Mạng nuôi dưỡng tái phát (RNN) nhiều bước với mô hình Bộ nhớ ngắn hạn dài hạn (LSTM) tạo ra một điểm dự đoán duy nhất về lưu lượng truy cập 2 giờ trong tương lai, với cửa sổ 6 giờ trước đó.

## Issues for traffic congestion: Các vấn đề tắc nghiễn giao thông

Ở nhiều nơi trên thế giới, tắc nghẽn giao thông là một vấn đề nghiêm trọng. Rất khó để xác định và dự báo chính xác vì nó bị ảnh hưởng bởi một loạt các yếu tố đa dạng. Một số vấn đề chính gây ra bởi tắc nghẽn giao thông là ô nhiễm không khí, đi lại dài, ùn tắc giao thông, giao thông công cộng không đáng tin cậy và tai nạn giao thông.
Để tạo điều kiện lưu lượng giao thông thông suốt trên mạng lưới đường bộ, chúng ta phải giải quyết vấn đề chính là tắc nghẽn giao thông. Do nhu cầu giao thông tăng cao, việc mở rộng cơ sở hạ tầng mạng lưới đường bộ bằng cách mở rộng đường bộ đơn giản là không đủ để quản lý các điều kiện lưu lượng giao thông thông suốt. Để mô hình hóa các trường hợp lưu lượng giao thông, một số loại phương pháp mô hình hóa lưu lượng giao thông là cần thiết mà chúng tôi sẽ cố gắng thực hiện trong dự án này.

## Intelligent transportation system (ITS): Hệ thống giao thoogn thông minh ITS

ITS là một hệ thống quản lý giao thông tích hợp bao gồm các công nghệ truyền thông dữ liệu, xử lý thông tin và quản lý giao thông tiên tiến. Trong những năm gần đây, sự thành công của học sâu trong thị giác máy tính, nhận dạng tốc độ và xử lý ngôn ngữ tự nhiên khiến việc áp dụng nó vào ITS trở nên tự nhiên. Chúng tôi chia các ứng dụng trong ITS thành các nhiệm vụ nhận dạng hình ảnh, TFP, dự đoán tốc độ giao thông (TSP), dự đoán thời gian di chuyển (TTP) và các nhiệm vụ khác. Thuật ngữ 'ITS' đề cập đến việc sử dụng các hệ thống liên lạc, thông tin, giao thông vận tải và giao thông đô thị. Hai trong số các mục tiêu chính của ITS là hiệu quả và an toàn giao thông. Giảm tình trạng tắc nghẽn và chậm trễ của giao lộ, cải thiện thời gian giao thông, cải thiện kiểm soát tốc độ, quản lý năng lực và quản lý sự cố là tất cả những lợi ích của ITS. Hình 1 mô tả nhiều nhiệm vụ được bao phủ bởi ITS.

## Data types, sources and TFP parameters:Kiểu dữ liệu, nguồn và tham số TFP:

Lưu lượng truy cập bị ảnh hưởng bởi nhiều yếu tố mà chúng ta cần xem xét để có dự đoán chính xác. Vì vậy, có một số nhóm dữ liệu chính mà chúng ta sẽ phải thu thập.   
**Mapping data:**  
Trước hết, chúng ta cần có một bản đồ chi tiết về mạng lưới đường bộ với các thuộc tính liên quan. Kết nối với các nhà cung cấp dữ liệu bản đồ toàn cầu như Google Maps, TomTom, HERE hoặc OSM là một cách tuyệt vời để có được thông tin đầy đủ và cập nhật.
**Traffic information:**  
Sau đó, chúng ta cần thu thập cả thông tin liên quan đến giao thông trong lịch sử và hiện tại như số lượng phương tiện đi qua tại một điểm nhất định, tốc độ và loại của chúng (xe tải, xe hạng nhẹ, v.v.). Các thiết bị được sử dụng để thu thập dữ liệu này là máy dò vòng lặp, máy ảnh, cảm biến chuyển động cân và radar hoặc các công nghệ cảm biến khác. May mắn thay, chúng tôi không phải tự mình cài đặt các thiết bị này ở khắp mọi nơi. Việc lấy thông tin này từ các nhà cung cấp nói trên thu thập dữ liệu từ hệ thống cảm biến, nguồn của bên thứ ba đa dạng hoặc sử dụng dữ liệu thăm dò GPS sẽ dễ dàng hơn. Các nền tảng khác như Otonomo sử dụng công nghệ Vehicle to Everything (V2X) sáng tạo để thu thập cái gọi là dữ liệu xe hơi được kết nối từ modem nhúng. Chúng tôi cũng có thể nhận được thông tin quan trọng khác về các sự cố (đóng đường hoặc công trình đường bộ), địa điểm yêu thích, v.v., từ các nhà cung cấp dữ liệu.
**Weather information:**  
Dữ liệu thời tiết (lịch sử, hiện tại và dự báo) cũng cần thiết vì điều kiện khí tượng ảnh hưởng đến tình hình đường và tốc độ lái xe. Có rất nhiều nhà cung cấp dữ liệu thời tiết như OpenWeather hoặc Tomorrow.io. 
**Additional data on road conditions:**
Có các nguồn dữ liệu bên ngoài có thể cung cấp thông tin quan trọng ảnh hưởng đến giao thông như bài đăng trên phương tiện truyền thông xã hội, tin tức địa phương hoặc thậm chí là máy quét của cảnh sát.
Nói tóm lại, để dự đoán TFP, chúng tôi xem xét phương pháp dự đoán đa thông số có tính đến các mẫu lưu lượng truy cập theo nhiều cách khác nhau.

 
**a) Flow:**  
 Lượng giao thông đi qua một điểm cụ thể trên đường trong một khoảng thời gian nhất định được gọi là luồng giao thông(the flow of traffic) . 
**b) Speed:**  
Khoảng cách di chuyển trên một đơn vị thời gian xác định tốc độ của xe. Trong hầu hết các trường hợp, tốc độ của bất kỳ phương tiện nào trên đường sẽ khác với những phương tiện khác xung quanh do các yếu tố như vị trí của người lái và điều kiện giao thông. 
**c) Day:**  
Ngày có thể là Chủ nhật đến Thứ Bảy 
**d) Day of type:**  
Ngày của loại chủ yếu được mô tả là ngày lễ, cuối tuần và làm việc.
**e) Clock time:**  
Thời gian đồng hồ có thể được chia thành giờ, tổng cộng là 24 giờ (1- 24 giờ). 
**f) Weather condition:**  
Dữ liệu thời tiết như nắng và mưa có thể được đưa đến mục đích đào tạo và diệt vong.

## Traffic theory: Lý thuyêt giao thông

Trong phần này, các biến lưu lượng truy cập và mật độ được giới thiệu. Khái niệm về dòng chảy tự do cũng được giải thích. Lưu lượng giao thông(Traffic flow): Lưu lượng giao thông (q) được định nghĩa là số lượng phương tiện trên một đơn vị thời gian trong một điểm tham chiếu.   
$q=n/∆T$  

**Traffic density:** Mật độ giao thông: Mật độ giao thông (k) là số lượng phương tiện có mặt trên mỗi chiều dài của đường. Đối với một con đường nhất định, điều này có nghĩa là chúng ta có $k_c$ dđược định nghĩa là mật độ tới hạn. Đây là lúc đường có lưu lượng giao thông cao điểm. Khi mật độ đi lên cao hơn, lưu lượng giao thông sẽ giảm và đến một lúc nào đó đạt mật độ ùn tắc $k_j$. Sau đó tắc nghẽn tối đa đã đạt đến đỉnh điểm.

## Literature Review: đánh giá

Cuộc khảo sát tài liệu về TFP và ITS được thực hiện bằng cách sử dụng một số nguồn thông tin thứ cấp như bài báo, Sách và Báo cáo nghiên cứu được xuất bản trong các ấn phẩm khác nhau. Thông tin thu thập được sau đó được xem xét để khám phá bất kỳ lĩnh vực quan tâm quan trọng nào của TFP và ITS. 

Việc dự đoán tắc nghẽn giao thông, đặc biệt là dự báo giao thông ngắn hạn, được thực hiện bằng cách đánh giá các thông số giao thông khác nhau. Để dự đoán tắc nghẽn giao thông, hầu hết các nghiên cứu sử dụng dữ liệu lịch sử. Tuy nhiên, một số báo cáo đã đưa ra dự đoán thời gian thực về tắc nghẽn giao thông. Trong giao thông thành phố và quản lý khu vực, TFP có một loạt các ứng dụng. Vấn đề TFP là một vấn đề chuỗi thời gian (TS) liên quan đến việc ước tính lưu lượng giao thông đường bộ đô thị tại một thời điểm trong tương lai bằng cách sử dụng thông tin thu thập được từ một hoặc nhiều điểm quan sát trong các giai đoạn trước. Hơn nữa, để dự đoán lưu lượng giao thông, nhiều kỹ thuật AI đặc biệt là các mô hình ML và DL khác nhau được áp dụng. ITS có thể giảm tắc nghẽn giao thông một cách hiệu quả bằng cách dự báo lưu lượng giao thông ngắn hạn. 

Graph neural networks (GNNs) là các mô hình học sâu sử dụng dữ liệu đồ thị làm đầu vào và được sử dụng cho nhiều ứng dụng khác nhau, bao gồm dự đoán thuộc tính phân tử và dự báo lưu lượng truy cập. Google Maps sử dụng máy học để kết hợp các điều kiện giao thông hiện tại với các mẫu giao thông trong quá khứ cho các con đường trên toàn thế giới nhằm dự đoán giao thông trong tương lai một cách đáng tin cậy. Mục đích của dự báo lưu lượng truy cập là đưa ra dự đoán lưu lượng truy cập bằng cách sử dụng dữ liệu khối lượng và tốc độ trong quá khứ. Các cơ quan chức năng đang chú ý hơn đến việc quan sát ùn tắc giao thông do sự phát triển của ngành giao thông vận tải và việc thu thập số liệu thống kê giao thông. Dữ liệu luồng giao thông được chuyển đổi thành ma trận 2D để xử lý trong dự báo lưu lượng do khả năng xử lý hình ảnh vượt trội của CNN.

Trong những năm gần đây, các mô hình và khung dự đoán lưu lượng giao thông mới đã nhanh chóng được phát triển để tăng hiệu suất TFP ngoài việc áp dụng các kỹ thuật trí tuệ nhân tạo (AI) như học máy (ML). Phần lớn dữ liệu giao thông là quy mô nhỏ, tập trung vào giao thông đường cao tốc hoặc thiếu dữ liệu phụ trợ. Điều kiện lưu lượng có thể được dự đoán bằng cả dữ liệu trực tuyến và ngoại tuyến, đáng chú ý là tìm kiếm trên các ứng dụng MAP. 

B.Karthika et al. đã thảo luận về dữ liệu di động phong phú và học sâu về dự đoán giao thông đô thị, các thuật toán sâu để dự báo dữ liệu giao thông trong thế giới thực và khi dữ liệu giao thông trở thành dữ liệu lớn, một số kỹ thuật để cải thiện độ chính xác của dự đoán giao thông cũng được thảo luận [1].  
Apurv Chandel et al. đã sử dụng nhiều mô hình hồi quy tuyến tính cho tập dữ liệu và thu được kết quả tốt nhất trong sự tương ứng của hồi quy tăng cường gradient sau khi thực hiện điều chỉnh các siêu tham số với độ chính xác là 83% [2]. 
Rohit Singh et al. trong nghiên cứu của mình trình bày một số kỹ thuật học máy hồi quy kết hợp hiệu suất của các thuật toán khác nhau và so sánh nó với các Thuật toán học máy hiện có bằng cách sử dụng lỗi bình phương trung bình, lỗi bình phương trung bình gốc và các số liệu hiệu suất khác [6].
Renhe Jiang và cộng sự đã trình bày một cuộc khảo sát toàn diện về tất cả các mô hình học sâu được sử dụng để dự đoán lưu lượng truy cập và triển khai bộ dữ liệu bằng cách sử dụng luồng pytorch và tensor để dự đoán tốt hơn [7].
 Kanokwan Khiewwan et al. đã mô tả các kỹ thuật khai thác dữ liệu để triển khai và sử dụng DT, KNN và SVM để mô hình hóa và phát hiện ra rằng DT hoạt động tốt hơn các thuật toán khác với độ chính xác 79,9% [8].

# 2. Methodology: Phương pháp luận

## Algorithms for generating traffic predictions (MODELS):Thuật toán để tạo dự đoán lưu lượng (mô hình):

Dự đoán giao thông liên quan đến việc dự báo tốc độ lái xe trên các đoạn đường cụ thể, cũng như sự xuất hiện và tiến hóa của ùn tắc. Chúng ta hãy xem xét các cách tiếp cận khác nhau cho nhiệm vụ này.

**Statistical approach:**  
Phương pháp thống kê: Sử dụng các kỹ thuật thống kê, chúng ta có thể nhận ra các mẫu lưu lượng truy cập ở các quy mô khác nhau, chẳng hạn như trong ngày, vào các ngày khác nhau trong tuần, theo mùa, v.v. So với các phương pháp học máy, chúng thường nhanh hơn, rẻ hơn và dễ thực hiện hơn. Tuy nhiên, vì chúng không thể xử lý nhiều dữ liệu đa biến, chúng kém chính xác hơn. Kể từ những năm 1970, các mô hình trung bình động tích hợp tự động hồi quy (ARIMA) — dễ sử dụng và hiển thị độ chính xác cao hơn các kỹ thuật thống kê khác — đã được sử dụng rộng rãi để dự đoán lưu lượng truy cập. Nó sử dụng một phương pháp thống kê truyền thống để phân tích quá khứ và dự báo tương lai. Nó thu thập dữ liệu từ một loạt các khoảng thời gian đều đặn và đưa ra giả định rằng các mô hình lịch sử sẽ tiếp tục trong tương lai. Tuy nhiên, lưu lượng truy cập là một cấu trúc phức tạp với nhiều biến không thể được xử lý hiệu quả với sự trợ giúp của các mô hình ARIMA đơn biến.  

**Machine learning approach:**  
Phương pháp học máy: Chúng ta có thể xây dựng các mô hình dự đoán bằng cách sử dụng học máy (ML) có tính đến lượng lớn dữ liệu không đồng nhất từ nhiều nguồn. Việc sử dụng các thuật toán ML để dự báo lưu lượng truy cập đã là chủ đề của nhiều nghiên cứu. Dưới đây là một vài ví dụ hiệu quả. Phương pháp rừng ngẫu nhiên xây dựng một số cây quyết định và kết hợp dữ liệu của chúng để đưa ra dự báo chính xác. Với đủ dữ liệu đào tạo, nó có thể tạo ra kết quả hiệu quả một cách nhanh chóng. Trong trường hợp này, các biến đầu vào của mô hình bao gồm thời tiết, thời gian, điều kiện đường xá cụ thể, chất lượng đường và ngày lễ. Hơn nữa, phương pháp k-nearest neighbors (KNN) sử dụng ý tưởng về sự tương đồng tính năng để dự báo các giá trị trong tương lai.

**Deep learning approach:**  
Phương pháp học sâu: Các phương pháp học sâu (DL) đã chứng minh hiệu quả cao trong việc dự đoán giao thông đường bộ so với ML hoặc kỹ thuật thống kê, luôn cho thấy độ chính xác dự báo khoảng 90% và cao hơn. Các thuật toán DL dựa trên mạng lưới thần kinh. Mạng lưới thần kinh nhân tạo (ANN) hoặc mạng thần kinh (NN) được tạo thành từ các nút liên kết với nhau (tế bào thần kinh) được tổ chức thành hai hoặc nhiều lớp và nhằm bắt chước hành vi của bộ não con người. Có nhiều loại mạng thần kinh được phát triển cho các mục đích khác nhau. Dưới đây là một số đã được sử dụng trong phân tích và dự đoán lưu lượng truy cập. Mạng nơ-ron tích chập (CNN) được coi là những người tiên phong trong ngành trong phân tích và nhận dạng hình ảnh. Phát hiện tắc nghẽn bằng cách sử dụng hình ảnh từ camera giám sát trên đường là một trong những ứng dụng tự nhiên của nó đối với các vấn đề giao thông. CNN không phải là lựa chọn đầu tiên cho dự báo giao thông. Tuy nhiên, những nỗ lực phát triển các mô hình dựa trên CNN để dự đoán tốc độ mạng lưới giao thông có hiệu quả cao. Để làm điều này, các nhà nghiên cứu đã tạo ra một ma trận hình ảnh 2 chiều từ dữ liệu thời gian và không gian đặc trưng cho lưu lượng giao thông. Mạng nơ-ron tái phát (RNN), so với mạng nơ-ron tích chập (CNN), được thiết kế để phân tích dữ liệu chuỗi thời gian hoặc quan sát được thu thập trong các khoảng thời gian cụ thể. Những hiểu biết như vậy có thể được nhìn thấy tốt trong các mô hình lưu lượng truy cập. Các mô hình RNN đã được chứng minh là dự đoán sự phát triển tắc nghẽn với độ chính xác tuyệt vời. Vấn đề gradient biến mất, đó là lý do tại sao RNN được coi là 'có bộ nhớ ngắn hạn', là nhược điểm của nó vì nó khiến một số dữ liệu từ các lớp trước đó bị mất. Đào tạo mô hình khó khăn và tốn thời gian hơn do kết quả của sự 'hay quên' này. Bộ nhớ ngắn hạn dài (LSTM) và đơn vị tái phát có cổng (GRU) là các biến thể của RNN giải quyết vấn đề gradient biến mất. Một nghiên cứu so sánh hiệu suất của các mô hình này cho thấy mô hình GRU chính xác hơn trong dự đoán lưu lượng giao thông và dễ đào tạo hơn. Nhiều nghiên cứu đã khuyến nghị phát triển các mô hình NN khác nhau để dự đoán lưu lượng truy cập, bao gồm mạng nơ-ron đồ thị, NN mờ, NN Bayes và các mô hình khác, cũng như sử dụng các kỹ thuật lai tích hợp hai hoặc nhiều thuật toán. Hiện tại, không có một kỹ thuật lý tưởng duy nhất nào có thể được sử dụng trong mọi tình huống để tạo ra các phép chiếu chính xác nhất. Có một vài điều nữa cần đề cập liên quan đến việc triển khai các kỹ thuật ML để dự đoán lưu lượng truy cập. Chúng ta phải nhớ rằng các thuật toán ML \/ DL hoạt động tốt nhất khi có đủ dữ liệu để đào tạo các mô hình và tinh chỉnh chúng để đạt được độ chính xác tối đa. Vì vậy, các bộ dữ liệu lớn hơn chúng tôi quản lý để có được kết quả tốt hơn chúng tôi sẽ nhận được.

## Comparing the performance of our model with other novel algorithms:So sánh hiệu suất của mô hình của chúng tôi với các thuật toán mới khác

**LSTM:**  
LSTM: Dự đoán lưu lượng giao thông ngắn hạn là rất quan trọng đối với các hệ thống giao thông thông minh và thông tin giao thông không gian và thời gian bổ sung có thể được sử dụng trong các nghiên cứu trong tương lai để ước tính chính xác lưu lượng trên mạng lưới đường quy mô lớn hơn. Nhìn chung, LSTM càng có nhiều lớp, năng lực học tập của mô hình càng tốt, nhưng nó cũng dễ bị quá tải.

**CNN:**   
CNN, hoặc mạng nơ-ron tích chập, thêm các lớp 'lọc' bổ sung nơi trọng số bộ lọc được xác định. Truyền ngược vẫn đang thực hiện nhiệm vụ này cho chúng tôi, nhưng chúng tôi sẽ không làm cho nó quá dễ dàng cho động cơ đáng tin cậy là backprop. CNN có nhiều bộ lọc song song có thể được điều chỉnh để trích xuất các khía cạnh quan tâm khác nhau.

**CNN VS LSTM:**   
CNN có thể được sử dụng để giảm thiểu số lượng tham số cần thiết cho đào tạo trong khi vẫn duy trì hiệu suất - đây là sức mạnh của việc kết hợp xử lý tín hiệu với học sâu, trong khi LSTM yêu cầu nhiều tham số hơn CNN, lợi ích của nó xuất phát từ việc có thể kiểm tra các chuỗi đầu vào dài mà không mở rộng kích thước mạng.
**LSTM RNN:** 

Kiến trúc RNN truyền thống có cái gọi là vấn đề gradient biến mất. Để khắc phục nhược điểm đó, một số cấu trúc nhất định của RNN như LSTM đã được đề xuất, được thiết kế để cung cấp cho các tế bào nhớ khả năng xác định khi nào nên quên một số thông tin nhất định, do đó xác định độ trễ thời gian tối ưu cho các vấn đề chuỗi thời gian. Những tính năng này đặc biệt mong muốn để dự đoán lưu lượng giao thông ngắn hạn trong lĩnh vực vận tải vì khả năng bộ nhớ lâu dài của nó . Sử dụng LSTM RNN để dự đoán lưu lượng truy cập và cho thấy LSTM RNN có hiệu suất tốt hơn hầu hết các mô hình phi tham số.

#### WindowGenerator
Tôi đã sửa đổi lớp WindowGenerator để có tất cả các phương thức và thuộc tính cần thiết được hợp nhất từ việc tiếp nhận trái ngược với thiết kế ban đầu của việc thêm các phương thức và thuộc tính ở các bước tiếp theo.

```python
'''
[Mã được cung cấp bởi hướng dẫn]
Lớp WindowGenerator.
Bắt đầu bằng cách tạo lớp `windowgenerator`.Phương thức `__init__` bao gồm
Tất cả các logic cần thiết cho các chỉ số đầu vào và nhãn.
Nó cũng lấy các khung data của tàu, eval và thử nghiệm làm đầu vào.
Chúng sẽ được chuyển đổi thành `tf.data.dataset` sau này.

Lớp học này có thể:

1 - Xử lý các chỉ mục và độ lệch như trong các sơ đồ trên.
2 - Chia cửa sổ các tính năng thành các cặp (tính năng, nhãn).
3 - Vẽ nội dung của các cửa sổ kết quả.
4 - Tạo hiệu quả các lô của các cửa sổ này từ dữ liệu đào tạo, đánh giá và kiểm tra, sử dụng tf.data.datasets.
'''
class WindowGenerator():
  #khởi tạo đối tượng
  def __init__(self, input_width, label_width, shift, batch_size=32,
               train_df=scaled_train_df, val_df=scaled_val_df, test_df=scaled_test_df,
               label_columns=None):
    #Lưu trữ dữ liệu thô.
    self.train_df = train_df
    self.val_df = val_df
    self.test_df = test_df
    self.example = None

    # Làm việc ra các chỉ số cột nhãn.
    self.label_columns = label_columns
    if label_columns is not None:
      self.label_columns_indices = {name: i for i, name in
                                    enumerate(label_columns)}
    self.column_indices = {name: i for i, name in
                           enumerate(train_df.columns)}

    # Làm việc với window parameters.
    self.input_width = input_width
    self.label_width = label_width
    self.shift = shift

    self.total_window_size = input_width + shift

    self.input_slice = slice(0, input_width)
    self.input_indices = np.arange(self.total_window_size)[self.input_slice]

    self.label_start = self.total_window_size - self.label_width
    self.labels_slice = slice(self.label_start, None)
    self.label_indices = np.arange(self.total_window_size)[self.labels_slice]

    self.batch_size = batch_size

  def __repr__(self):
    return '\n'.join([
        f'Total window size: {self.total_window_size} = {self.input_width}+({self.shift-self.label_width}+{self.label_width})',
        f'Batch size: {self.batch_size}',
        f'Input indices: {self.input_indices}',
        f'Label indices: {self.label_indices} start: [{self.label_start}]',
        f'Label column name(s): {self.label_columns}'])

  def split_window(self, features):
    '''
      Cho một danh sách đầu vào liên tiếp, phương pháp split_window sẽ chuyển đổi
      chúng đến một cửa sổ đầu vào và một cửa sổ nhãn.
    '''
    inputs = features[:, self.input_slice, :]
    labels = features[:, self.labels_slice, :]
    if self.label_columns is not None:
      labels = tf.stack(
          [labels[:, :, self.column_indices[name]] for name in self.label_columns],
          axis=-1)
# Cắt không bảo toàn thông tin hình dạng tĩnh, vì vậy hãy đặt hình dạng
    # thủ công.Bằng cách này, `tf.data.datasets` dễ kiểm tra hơn.
    inputs.set_shape([None, self.input_width, None])
    labels.set_shape([None, self.label_width, None])

    return inputs, labels

  def plot(self, model=None, plot_col='traffic_volume', max_subplots=3):
    '''
      Phương pháp cốt truyện cho phép hình dung đơn giản của cửa sổ chia
    '''
    inputs, labels = self.example
    plt.figure(figsize=(12, 8))
    plot_col_index = self.column_indices[plot_col]
    max_n = min(max_subplots, len(inputs))
    for n in range(max_n):
      plt.subplot(max_n, 1, n+1)
      plt.ylabel(f'{plot_col} [normed]')
      plt.plot(self.input_indices, inputs[n, :, plot_col_index],
              label='Inputs', marker='.', zorder=-10)

      if self.label_columns:
        label_col_index = self.label_columns_indices.get(plot_col, None)
      else:
        label_col_index = plot_col_index

      if label_col_index is None:
        continue

      plt.scatter(self.label_indices, labels[n, :, label_col_index],
                  edgecolors='k', label='Labels', c='#2ca02c', s=64)
      if model is not None:
        predictions = model(inputs)
        plt.scatter(self.label_indices, predictions[n, :, label_col_index],
                    marker='X', edgecolors='k', label='Predictions',
                    c='#ff7f0e', s=64)

      if n == 0:
        plt.legend()

    plt.xlabel('Time [h]')
   # Maskar: Đã thêm dòng này cho thấy rằng chúng tôi chỉ hiển thị n lát
    plt.suptitle(f'Showing only {max_n} slices')

  def make_dataset(self, data):
    '''
     Phương pháp Make_DataSet sẽ lấy một chuỗi thời gian DataFrame và chuyển đổi nó thành
      một tf.data.dataset của (input_window, label_window)
      tiền xử lý.timeseries_dataset_from_array.
    '''
    data = np.array(data, dtype=np.float32)
    ds = tf.keras.preprocessing.timeseries_dataset_from_array(
        data=data,
        targets=None,
        sequence_length=self.total_window_size,
        sequence_stride=1,
        shuffle=False, # Maskar: I changed this to False
        batch_size=self.batch_size, # 32,
        )

    ds = ds.map(self.split_window)

    return ds

# Đối tượng WindowGenerator giữ dữ liệu đào tạo, xác thực và kiểm tra.
  # Thêm thuộc tính để truy cập chúng dưới dạng tf.data.dataSets bằng phương pháp make_dataset ở trên.
  # Ngoài ra, thêm một lô ví dụ tiêu chuẩn để dễ dàng truy cập và vẽ âm mưu:
  @property
  def train(self):
    return self.make_dataset(self.train_df)

  @property
  def val(self):
    return self.make_dataset(self.val_df)

  @property
  def test(self):
    return self.make_dataset(self.test_df)

  @property
  def example(self):
    """Get and cache an example batch of `inputs, labels` for plotting."""
    result = getattr(self, '_example', None)
    if result is None:
      # No example batch was found, so get one from the `.train` dataset
      result = next(iter(self.train))
      # Và lưu trữ nó cho lần sau
      self._example = result
    return result

  @example.setter
  def example(self, var):
    self._example = var
```

```python
def plot_train_validation(history, validation_data, metric='mean_absolute_error'):
    '''
    Function to plot the training and validation accuracy progress with epochs for a single model
    '''
    plt.plot(history.history[metric], label='train')  # For TF2
    plt.plot(history.history['val_'+metric], label='valid.')  # For TF2
    plt.xlabel('Epoch')
    plt.ylabel(metric)
    #plt.xticks(range(1,len(history.history['loss'])+1,2))
    #plt.ylim([0.5, 1])
    plt.title(f'{len(history.history[metric])} epochs')
    plt.grid(color='lightgrey', linestyle=':', linewidth=0.5)
    plt.legend(loc='upper right')
    plt.show()

    # Đánh giá mô hình đã học với bộ xác thực
    print(f'\n\n{StartBold}Validation Scores:{EndBold}\n')
    _ = history.model.evaluate(validation_data, verbose=2)
```

```python
class TrainingPlot(Callback):
    '''
     Chức năng trợ giúp gọi lại Keras
      Biểu đồ Tained/xác nhận sau mỗi kỷ nguyên
    '''
    def __init__(self, loss_metric='mean_absolute_error'):
      super().__init__()
      self.loss_metric = loss_metric

    # Chức năng này được gọi khi đào tạo bắt đầu
    def on_train_begin(self, logs={}):
        # Khởi tạo các danh sách để giữ các bản ghi, tổn thất và số liệu
        self.losses = []
        self.val_losses = []
        self.logs = []
    
    # Phương thức này được gọi khi kết thúc mỗi epoch.
    def on_epoch_end(self, epoch, logs={}):
        loss = logs.get(self.loss_metric)
        val_loss = logs.get('val_'+self.loss_metric)
                
        # Nối nhật ký, tổn thất và độ chính xác vào danh sách
        self.logs.append(logs)
        self.losses.append(loss)
        self.val_losses.append(val_loss)
        
        # Plots every n-th epoch
        if epoch > 0 and epoch%1==0:
            
            # Clear the previous plot
            IPython.display.clear_output(wait=True)
            #clear_output(wait=True)
            N = np.arange(0, len(self.losses))
            
            # You can chose the style of your preference
            #plt.style.use("seaborn")            
            #plt.figure(figsize=(10,3))
            plt.xlabel('Epoch')
            plt.ylabel(self.loss_metric)
            plt.plot(N, self.losses, label='train')
            plt.plot(N, self.val_losses, label='valid.')
            plt.grid(color='lightgrey', linestyle=':', linewidth=0.5)
            val_denorm = np.float(my_scaler_tv.inverse_transform(np.array([[val_loss]])))
            plt.title(f'Last validation value = {val_denorm:.2f}')
            plt.legend(loc='upper right')          
            plt.show()
```

### mô tả quá trình biên dịch và đào tạo

```python
def compile_and_fit(model, window, patience=5, max_epochs=30,
                    model_label='model', log_dict=None):
    plot_metrics = TrainingPlot()

    rlr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, min_delta=0.001, patience=patience, mode='min', verbose=1 )

    #mcp = ModelCheckpoint(filepath='weights.h5', monitor='val_loss', verbose=1, save_best_only=True, save_weights_only=True)

    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=patience, mode='min', restore_best_weights=True )

    model.compile( optimizer=tf.optimizers.Adam(), loss=tf.losses.MeanSquaredError(), metrics=[tf.metrics.MeanAbsoluteError()] )

    history = model.fit( window.train, epochs=max_epochs, validation_data=window.val, callbacks=[plot_metrics, early_stopping, rlr] )
  
    if log_dict:
        log_dict['my_models'][model_label] = model

        IPython.display.clear_output()

        print(f'\n\n{StartBold}Training vs. Validation:{EndBold}\n')
        plot_train_validation(history, window.val)

        log_dict['multi_val_performance'][model_label] = model.evaluate(window.val, verbose=0)
        log_dict['multi_performance'][model_label] = model.evaluate(window.train, verbose=0)


    predictions = model.predict(window.val)
    evaluate_predictions(val_df.traffic_volume, predictions)  
    return history
```

### Đánh giá dự đoán

```python
def evaluate_predictions(y_true, y_pred, plot_start_index=-500):
    print(f'\n\n{StartBold}Predictions Evaluation:{EndBold}\n')
    y_p = np.asarray(my_scaler_tv.inverse_transform(y_pred.reshape(-1,1))).ravel()
    n_predictions = len(y_p)
    y = np.asarray(y_true[-n_predictions:]).ravel()
    print('Predictions:', n_predictions)
    mae = np.float(tf.metrics.mean_absolute_error(y, y_p))
    mae_scaled = np.float(my_scaler_tv.transform(np.array([[mae]])))
    print(f'MAE: {mae:.2f} ({mae_scaled:.4f})')

    plt.subplots(figsize=(15,2))
    plt.plot(y[plot_start_index:], marker='.', label='true')
    plt.plot(y_p[plot_start_index:], marker='.', label='predicted')
    plt.legend()
    plt.show()
```
